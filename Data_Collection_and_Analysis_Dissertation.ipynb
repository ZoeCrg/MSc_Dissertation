```python
def read_the_csv(file):
    df = pd.read_csv(file)
    return df

def parse_material_compositions(teststring, recycled_tag='Recycled'):
    try:
        splitup = (teststring.split(","))
    except:
        return [[],[],[]]
    lists = []
    newlist = []
    for i in splitup:
        i = i.strip()
        if len(i) == 0:
            continue
        if i[0] == "[":
            try:
                if newlist:
                    lists.append(newlist)
            except: 
                pass
            newlist = []
        newlist.append(i)
    lists.append(newlist)
    for l in lists:
        try:
            b = l[0].split('] ')
            l[0] = b[1]
        
            if b[0] == "[Recycled":  
                l.insert(0, "Recycled")
        except: 
            pass
    return lists

def get_recycled(lists):
    for i in lists:
        try:
            if(i[0] == "Recycled"):
                return i[1:]
        except:
            pass

def get_main(lists):
    return lists[0]

def get_secondary(lists):
    if len(lists) <2:
        return
    secondary = []
    for i in lists[1:]:
        if(i[0] == "Recycled"):
            continue
        secondary.append(i)
    if len(secondary) == 0:
        return
    return secondary

def remove_trademark_symbol(input_string):
    # Define a regular expression pattern to match the trademark symbol (\u2122 and \U00002122) in a case-insensitive manner
    pattern = r'\\[Uu]([0-9a-fA-F]{4,8})'

    # Use re.sub() to replace all occurrences of the pattern with an empty string
    cleaned_string = re.sub(pattern, '', input_string)

    return cleaned_string

def split_material_and_percentage(input_string):
    # Define a regular expression pattern to match the material and percentage parts
    pattern = r'^(.*?)(\d+\%)$'
    
    # Use the re.match function to find the pattern in the input_string
    match = re.match(pattern, input_string)
    
    if match:
        # The first group (index 1) contains the material part
        material = match.group(1).strip()
        
        # The second group (index 2) contains the percentage part
        percentage = match.group(2).strip()
        material = remove_trademark_symbol(material)
        return material, percentage
    else:
        # If no match is found, return the entire string as material and an empty string for percentage
        return input_string.strip(), ""

def add_columns_and_values(df):
    # Iterate through each row of the DataFrame
    for index, row in df.iterrows():
        composition_list = get_main((parse_material_compositions(row["fabric"])))
        print(composition_list)
        # Extracting the material name and percentage from the list
        for item in composition_list:
            try:
                material, percentage = split_material_and_percentage(item)
                percentage = float(percentage.strip('%'))
                # Adding a new column to the DataFrame if it doesn't exist
                if material.upper() not in df.columns:
                    df[material.upper()] = 0  # Initializing the column with None
                # Assigning the percentage value to the corresponding cell
                df.at[index, material.upper()] = percentage
            except:
                #old fabric function may of failed so we will try the link again to get the materials
                link = str(row['link'])
                new_fabric_value = fabric_function(link)
                df.at[index, 'fabric'] = new_fabric_value 
                composition_list = get_main((parse_material_compositions(new_fabric_value)))
                # Extracting the material name and percentage from the list
                for item in composition_list:
                    try:
                        material, percentage = split_material_and_percentage(item)
                        percentage = float(percentage.strip('%'))
                        # Adding a new column to the DataFrame if it doesn't exist
                        if material.upper() not in df.columns:
                            df[material.upper()] = 0  # Initializing the column with None
                        # Assigning the percentage value to the corresponding cell
                        df.at[index, material.upper()] = percentage
                    except:
                        print('failed again')
                        pass
                
                
                pass
            
    
    return df


def add_recycled_columns_and_values(df):
    # Iterate through each row of the DataFrame
    for index, row in df.iterrows():
        composition_list = get_recycled((parse_material_compositions(row["fabric"])))
        # Extracting the material name and percentage from the list
        if composition_list:
            for item in composition_list:
                try:
                    
                    material, percentage = split_material_and_percentage(item)
                    
                    material = 'Sustainable ' + material
                    percentage = float(percentage.strip('%'))
                    # Adding a new column to the DataFrame if it doesn't exist
                    if material.upper() not in df.columns:
                        df[material.upper()] = 0  # Initializing the column with None
                    # Assigning the percentage value to the corresponding cell
                    df.at[index, material.upper()] = percentage
                except:
                    pass
    
    return df

def process_build_file(filepath):
    df=read_the_csv(filepath)
    materialsdf =add_columns_and_values(df)
    recycledmaterialsdf =add_recycled_columns_and_values(materialsdf)
    filename = os.path.basename(filepath)[:-4]
    recycledmaterialsdf['Date'] = filename
    return recycledmaterialsdf;

```

```python
def find_most_frequent_materials(df):
    # Get a list of all material columns (excluding non-material columns)
    material_columns = df.columns[df.columns.str.isupper()].tolist()

    # Create a dictionary to store the occurrences of each material
    materials_count = {}

    # Iterate through each row and count the occurrences of each material
    for _, row in df.iterrows():
        for material_column in material_columns:
            material_name = material_column.lower()
            material_value = row[material_column]
            if material_value != 0:
                materials_count[material_name] = materials_count.get(material_name, 0) + 1

    # Sort the materials by their occurrence count in descending order
    sorted_materials = sorted(materials_count.items(), key=lambda x: x[1], reverse=True)

    # Print out the table with the materials and their frequencies
    print("Most frequently used materials:")
    print("-----------------------------")
    print("Material\tOccurrences")
    print("-----------------------------")
    for material, count in sorted_materials:
        print(f"{material}\t\t{count}")

# Call the function with the provided DataFrame 'df'
find_most_frequent_materials(df)
```

```python
import plotly.graph_objects as go
import ipywidgets as widgets
from IPython.display import display

df = read_the_csv('august_data_example.csv')

def count_recycled_fabric_percentage_ranges(df, main_fabric):
    # Filter the DataFrame to include only rows with the specified main fabric in the material column
    df_main_fabric = df[df[main_fabric] > 0]
    recycled_columns = get_sustainable_column(df, main_fabric)
    # Initialize a dictionary to store the counts for each percentage range
    percentage_counts = {}

    # Define percentage range intervals (0%, 1-10%, 11-20%, ..., 91-100%)
    percentage_ranges = [(0, "0%")] + [(i, f"{i}-{i+9}%") for i in range(1, 91, 10)] + [(91, "91-100%")]
    
    for percentage_range, range_label in percentage_ranges:
        percentage_counts[range_label] = 0
    
    # Iterate through each row and count the occurrences in each percentage range
    for _, row in df_main_fabric.iterrows():
        recycled_fabric_content = 0
        for recycled_fabric in recycled_columns:
            recycled_fabric_content += row[recycled_fabric]
        for percentage_range, range_label in percentage_ranges:
            if percentage_range == 0:
                if recycled_fabric_content == 0:
                    percentage_counts[range_label] += 1
                    break
            elif percentage_range <= recycled_fabric_content < percentage_range + 10:
                percentage_counts[range_label] += 1
                break
    return percentage_counts

def get_sustainable_column(df, main_fabric):
    sustainable_columns = [col for col in df.columns if "SUSTAINABLE" in col]
    filtered_list = [string for string in sustainable_columns if main_fabric in string]
    return filtered_list

def get_table_data(df):
    # Get the list of main fabrics automatically from the DataFrame columns (exclude columns before 'fabric')
    main_fabrics = [
    "POLYESTER",
    "VISCOSE",
    "ELASTANE",
    "LYOCELL",
    "COTTON",
    "POLYAMIDE",
    "LINEN",
    "ACRYLIC",
    "METALLISED FIBRE",
    "WOOL",
    "MODAL",
    "CASHMERE",
    "POLYESTER RESIN",
    "ALPACA",
    "MOHAIR",
    "LEATHER",
    "POLYURETHANE",
    "ACETATE",
    "SILK",
    "RAMIE",
    "POLYSTYRENE FOAM",
    "POLYETHYLENE"]

    data = []
    for fabric in main_fabrics:
        percentages = count_recycled_fabric_percentage_ranges(df, fabric)
        data.append((fabric, percentages))

    
    headers = ["Material"] + list(data[0][1].keys())  # Get the percentage ranges as headers
    
    table_data = []
    for material, percentages in data:
        row = [material] + list(percentages.values())
        table_data.append(row)

    sorted_table_data = sorted(table_data, key=lambda row: sum(row[1:]), reverse=True)
    sorted_table_data.insert(0, headers)
    return sorted_table_data
   
def draw_table(df):
    data = get_table_data(df)
#     print(data)
    from tabulate import tabulate

    # Generate and display an ASCII table
    print(tabulate(data[1:], data[0], tablefmt="pretty"))

draw_table(df)
```

What are the most common blends of fibres?

```python
def get_non_sus_df(df):
    sustainable_col = []
    for column_name in df.columns:
        if column_name.startswith("SUSTAINABLE"):
             sustainable_col.append(column_name)
    
    selected_rows = []
    
    for _, row in df.iterrows():
        total_sustainable = sum(row[column_name] for column_name in sustainable_col)
        if total_sustainable == 0:
                selected_rows.append(row)
    
    # Create a new DataFrame from the selected rows
    non_sustainable_df = pd.DataFrame(selected_rows)
    return non_sustainable_df
def get_blends(df):
    df = get_non_sus_df(df)
    blends = {}

    main_fabrics = [
    "POLYESTER",
    "VISCOSE",
    "ELASTANE",
    "LYOCELL",
    "COTTON",
    "POLYAMIDE",
    "LINEN",
    "ACRYLIC",
    "METALLISED FIBRE",
    "WOOL",
    "MODAL",
    "CASHMERE",
    "POLYESTER RESIN",
    "ALPACA",
    "MOHAIR",
    "LEATHER",
    "POLYURETHANE",
    "ACETATE",
    "SILK",
    "RAMIE",
    "POLYSTYRENE FOAM",
    "POLYETHYLENE"]

    for _, row in df.iterrows():
        materials_combo = []
        for material in main_fabrics:
            if row[material] > 0:
                materials_combo.append(material)

        combo = "+".join(materials_combo)

        value = blends.get(combo)

        if value is None:
            blends[combo] = 1
        else:
            blends[combo] = value + 1
                
    return blends
    
get_blends(df)

def print_keys_in_order_of_values(input_dict):
    # Sort the dictionary items by values in ascending order
    sorted_items = sorted(input_dict.items(), key=lambda x: x[1], reverse=True)
    total = 0
    total_blend_3_or_more = 0
    # Extract and print the keys in sorted order
    for key, value in sorted_items:
        print(key + ' ' + str(value))
        count = 0

        for char in key:
            if char == '+':
                count += 1
        if count > 1:
            total_blend_3_or_more += value
        total += value
    print('total items with no sustainable mat ('+str(total)+') and 3 or more blends: '+ str(total_blend_3_or_more))
print_keys_in_order_of_values(get_blends(df))
```
